Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Answer 1. Matplotlib is a Python library used for creating high-quality data visualizations in the form of charts, plots, and graphs. It is widely used for data analysis, scientific computing, and machine learning.

Matplotlib's popularity comes from its ability to create static, interactive, and animated visualizations in Python. It provides a flexible framework for creating visualizations and allows users to customize their charts and plots to meet their specific needs.

Pyplot is a sublibrary of Matplotlib that provides a simple interface for creating plots and charts.

Here are five types of plots that can be plotted using Pyplot module of Matplotlib:

1. Line plots: Line plots are used to display data points connected by straight lines. They are used to represent trends or patterns in the data over time.

2. Scatter plots: Scatter plots are used to display the relationship between two variables. They are used to identify clusters or patterns in the data.

3. Bar charts: Bar charts are used to compare categorical data across different groups or categories. They are used to represent data as bars of different heights or lengths.

4. Histograms: Histograms are used to display the distribution of data across different ranges or intervals. They are used to represent data as bars of different heights or lengths.

5. Pie charts: Pie charts are used to display the proportion of different categories or groups in the data. They are used to represent data as slices of a pie, with each slice representing a different category or group.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

Answer 2. A scatter plot is a type of plot used to visualize the relationship between two variables. It displays data points as a collection of points, where the position of each point is determined by its values for the two variables being plotted.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y, color= '#13BC2C')

plt.title("Sactter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Answer 3. The subplot() function is used to create multiple plots within the same figure. It takes three arguments: the number of rows, the number of columns, and the index of the plot. This function is useful when we want to visualize multiple plots side by side or in a grid layout.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0,1,2,3,4,5])
y1 = np.array([0,100,200,300,400,500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])


profit = np.array([3000, 8000, 1000, 10000])

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.title('Company Profit')
plt.xlabel("Company")
plt.ylabel("Profit")

plt.show()
```

```python
#Horizontal bar plot

plt.barh(company, profit)
plt.title('Company Profit')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

Answer 5. A box plot is a type of chart that displays the distribution of a dataset by showing the median, quartiles, and outliers. It is used to visualize the spread and skewness of a dataset, and to compare the distribution of multiple datasets.

A box plot consists of a rectangular box that represents the interquartile range (IQR), which contains the middle 50% of the data, with the median represented by a horizontal line inside the box. The "whiskers" of the box plot extend to the minimum and maximum values that are not considered outliers, and any data points outside the whiskers are shown as individual points, which are considered outliers.


```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
plt.boxplot(data)
plt.title('Box Plot')
plt.xticks([1,2], ['Box 1', 'Box 2'])
plt.ylabel('Value')


plt.show()
```
